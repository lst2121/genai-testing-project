```python
def weird_or_not(n):
    if n % 2 != 0:
        return "Weird"
    elif 2 <= n <= 5:
        return "Not Weird"
    elif 6 <= n <= 20:
        return "Weird"
    return "Not Weird"

print(weird_or_not(22))
```

```python
def print_squares(n):
    for i in range(n):
        print(i ** 2)

def get_sqrs(n):
    return [i ** 2 for i in range(n)]

print_squares(5)
print(get_sqrs(9))
```

```python
def print_1_to_n(n):
    print(*range(1, n+1), sep="")

print_1_to_n(3)

def get_1_to_n(n):
    return "".join(str(i) for i in range(1, n+1))

print(get_1_to_n(9))


```

```python
def list_comprehension_coords(x, y, z, n):
    return [
        [i, j, k]
        for i in range(x + 1)
        for j in range(y + 1)
        for k in range(z + 1)
        if i + j + k != n
    ]

list_comprehension_coords(1, 1, 1, 2)
```

```python
def runner_up_score(scores):
    unique = sorted(set(scores), reverse=True)
    return unique[1]

def runner_up_score_optimal(scores):
    first = second = float('-inf')
    
    for score in scores:
        if score > first:
            second = first
            first = score
        elif score > second and score != first:
            second = score
    
    return second

print(runner_up_score([2, 3, 6, 6, 5]))
print(runner_up_score_optimal([2, 3, 6, 6, 5]))


```

```python
def second_lowest_students(records):
    grades = sorted((set(grade for name, grade in records)))
    second_lowest = grades[1]

    students = sorted(name for name, grade in records if grade == second_lowest)
    return students


records = [("Harry", 37.21), ("Berry", 37.21), ("Tina", 37.2), 
               ("Akriti", 41), ("Harsh", 39)]
print(second_lowest_students(records))
```
