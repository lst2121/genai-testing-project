```python
def reverse_string(str):
    return str[::-1]

print(reverse_string("Hello"))
```

```python
def reverse_string_loop(str):
    result = []

    for char in str:
        result.insert(0, char)
    return "".join(result)

print(reverse_string_loop("flash"))
```

```python
def is_palindrome(text):
    cleaned = "".join([char.lower() for char in text if char.isalnum()])
    return cleaned == cleaned[::-1]

is_palindrome("Hello! Trump")
is_palindrome("A man, a plan, a canal: Panama")
```

```python
def is_palindrome_two_pointers(text):
    left, right = 0, len(text) - 1

    while left < right:
        while left < right and not text[left].isalnum():
            left += 1
        while left < right and not text[right].isalnum():
            right -= 1
        if text[left].lower() != text[right].lower():
            return False

        left += 1
        right -= 1
        
    return True
```

```python
is_palindrome_two_pointers("Hello! Trump")
```

```python

is_palindrome_two_pointers("A man, a plan, a canal: Panama")
```

```python
def is_palindrome_simple(text):
    cleaned = ""

    for char in text:
        if char.isalnum():
            cleaned += char.lower()
    
    reversed_text = ""

    for char in cleaned:
        reversed_text = char + reversed_text
    
    return cleaned == reversed_text

is_palindrome_simple("Hello Trump")
```

```python
def char_frequency(text):
    freq = {}
    text = text.lower()
    for char in text:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq

char_frequency("hello")
```

```python
def first_repeating_char(text):
    seen = []

    for char in text.lower():
        if char not in seen:
            seen.append(char)
        else:
            return char
    return None

first_repeating_char("Hell")
```

```python
def is_anagram(first, second):
    if len(first) != len(second):
        return False

    def frequency_map(text):
        freq = {}
        for char in text:
            freq[char] = freq.get(char, 0) + 1
        return freq
    return frequency_map(first) == frequency_map(second)

    
```

```python
is_anagram("listen", "silent")
```

```python
is_anagram("listen", "silents")
```

```python
def is_anagram_sorting(first, second):
    return sorted(first) == sorted(second)

is_anagram_sorting("listen", "silens")   
```

```python
def is_anagram_remove(first, second):
    if len(first) != len(second):
        return False
    second_chars = list(second)
    print(second_chars)

    for char in first:
        if char in second_chars:
            second_chars.remove(char)
        else:
            return False
    
    return len(second_chars) == 0

is_anagram_remove("listen", "silens")   
```

```python
def remove_duplicates_preserve_order(items):

    seen = set()
    result = []

    for item in items:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

remove_duplicates_preserve_order([1, 4, 2, 2, 3, 1, 5,7, 6, 7])
```

```python
def find_duplicates(items):
    duplicates = set()
    seen = set()

    for item in items:
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return(duplicates, seen)

find_duplicates([1, 2, 2, 3, 3, 3] )
```

```python
def common_elements(first, second):
    first_unq = set(first)
    second_unq = set(second)

    print(first_unq, second_unq)

    return list(first_unq & second_unq)

common_elements([1, 2, 3], [2, 3, 4] )
```

```python
def second_largest(nums):
    nums_unique = set(nums)
    nums_sorted = sorted(nums_unique, reverse=True)
    return nums_sorted[1]
second_largest([10, 5, 20, 20])
```

```python
def second_largest(nums: list[int]) -> int | None:
    unique_nums = sorted(set(nums), reverse=True)
    if len(unique_nums) < 2:
        return None
    return unique_nums[1]
second_largest([10, 5, 20, 20])
```

```python
def second_largest_single_pass(nums):
    first = second = None

    for num in nums:
        if num == first or  num ==second:
            continue

        if first is None or num > first:
            second = first
            first = num
        elif second is None or num > second:
            second = num
    
    return second


second_largest_single_pass([10, 5, 20, 20])
```

```python
def move_zeroes(nums):
    non_zeros = [num for num in nums if num != 0]
    print(non_zeros)
    zero_count = len(nums) - len(non_zeros)
    return non_zeros + [0] * zero_count

move_zeroes([0, 1, 0, 3, 12] )
```

```python
def move_zeroes_in_place(nums: list[int]) -> None:
    insert_position = 0

    for num in nums:
        if num != 0:
            nums[insert_position] = num
            insert_position += 1

    while insert_position < len(nums):
        nums[insert_position] = 0
        insert_position += 1

move_zeroes_in_place([0, 1, 0, 3, 12] )
```

```python
def move_zeroes_simple(nums):
    result = []
    zeros = []

    for num in nums:
        if num == 0:
            zeros.append(num)
        else:
            result.append(num)
    return result + zeros

move_zeroes_simple([0, 1, 0, 3, 12] )
```
